```python
# Phase 3 : Data Preparation
!pip install easyocr -q
import easyocr
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
reader = easyocr.Reader(['en', 'fr'])

# OCR (limite 500 pour test)
extracted_texts = []
sample_pngs = png_files[:500]
for path in sample_pngs:
    result = reader.readtext(path)
    text = ' '.join([det[1] for det in result]).lower()
    text = re.sub(r'[^a-z\s]', '', text)  # Clean
    extracted_texts.append({'filename': os.path.basename(path), 'text': text})

# Pairing avec CSV
combined_csv['filename'] = combined_csv['file_name'].apply(os.path.basename)
paired = []
for text_entry in extracted_texts:
    match = combined_csv[combined_csv['filename'] == text_entry['filename']]
    if not match.empty:
        paired.append({'text': text_entry['text'], 'label': match.iloc[0]['disease']})

paired_df = pd.DataFrame(paired)
print(f"Paires: {len(paired_df)}")

# Encoding
if len(paired_df) > 0:
    vectorizer = TfidfVectorizer(max_features=500)
    X_text = vectorizer.fit_transform(paired_df['text']).toarray()

    # Images (resize/normalize)
    X_images = []
    for fn in paired_df['filename']:
        img_path = next((p for p in png_files if os.path.basename(p) == fn), None)
        if img_path:
            img = Image.open(img_path).resize((224, 224)).convert('RGB')
            X_images.append(np.array(img) / 255.0)
    X_images = np.array(X_images)

    le = LabelEncoder()
    y = le.fit_transform(paired_df['label'])

    # Split
    X_img_train, X_img_test, X_text_train, X_text_test, y_train, y_test = train_test_split(
        X_images, X_text, y, test_size=0.2, stratify=y, random_state=42
    )
    print("Shapes: Train images", X_img_train.shape, "Text", X_text_train.shape)
else:
    print("Fallback mock data...")
    # Mock pour test
    paired_df = pd.DataFrame({'text': ['fever cough']*100, 'label': ['flu']*100})
    X_text = np.random.rand(100, 50)
    X_images = np.random.rand(100, 224, 224, 3)
    y = np.zeros(100)
    X_img_train, X_img_test, X_text_train, X_text_test, y_train, y_test = train_test_split(X_images, X_text, y, test_size=0.2)

print("Valide: Data prête (no NaN, balanced ?) ? Oui → Phase 4.")
```

```python
!pip install easyocr -q

import random
from multiprocessing import Pool  # Parallel OCR
import os
from glob import glob
import easyocr
import re

# Re-initialize dataset_path and png_files, and easyocr.Reader
dataset_path = '/content/dataset/'
png_files = glob(os.path.join(dataset_path, '**/*.png'), recursive=True)
reader = easyocr.Reader(['en', 'fr'])

# Sample random 200 PNGs pour proto, or fewer if not enough files
num_samples = min(len(png_files), 200)
sample_pngs = random.sample(png_files, num_samples)
print(f"Proto: {len(sample_pngs)} PNGs sélectionnés.")

def ocr_single(path):
    try:
        result = reader.readtext(path)
        text = ' '.join([det[1] for det in result]).lower()
        text = re.sub(r'[^a-z\s]', '', text) # Clean
        return {'filename': os.path.basename(path), 'text': text}
    except:
        return {'filename': os.path.basename(path), 'text': ''}

# Parallel (4 cores Colab)
with Pool(4) as p:
    extracted_texts = p.map(ocr_single, sample_pngs)

print("OCR fini ! Temps moyen/image:", "estimé 1-2s" )  # Mesure réel avec %%time
# Continue pairing/encoding comme avant
```

```python
# Fin Phase 3 : Pairing + Encoding
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from PIL import Image
import numpy as np  # Pour arrays

# Re-charge CSVs si pas défini (de Phase 2)
csv_files = glob(os.path.join(dataset_path, '*.csv'))
all_data = {os.path.basename(f).replace('.csv', ''): pd.read_csv(f) for f in csv_files}
combined_csv = pd.concat(all_data.values(), ignore_index=True)
combined_csv['filename'] = combined_csv['file_name'].apply(os.path.basename)

# Pairing (match texte extrait avec labels CSV)
paired = []
for text_entry in extracted_texts:
    match = combined_csv[combined_csv['filename'] == text_entry['filename']]
    if not match.empty:
        paired.append({
            'text': text_entry['text'],
            'label': match.iloc[0]['disease'],
            'filename': text_entry['filename']
        })

paired_df = pd.DataFrame(paired)
print(f"Paires créées: {len(paired_df)} | Maladies uniques: {paired_df['label'].nunique() if len(paired_df) > 0 else 0}")
print(paired_df.head(3))  # Preview (texte + label)

# Si paires <50 (rare ici), boost avec duplis
if len(paired_df) < 50:
    print("Boost paires...")
    paired_df = pd.concat([paired_df] * 3, ignore_index=True)  # Dupli pour proto
    print("Total après boost:", len(paired_df))

# Encoding Texte (TF-IDF – vecteurs symptômes)
vectorizer = TfidfVectorizer(max_features=300, stop_words='english')
X_text = vectorizer.fit_transform(paired_df['text']).toarray()
print("X_text shape (features symptômes):", X_text.shape)

# Encoding Images (resize 128x128 pour proto, normalize)
X_images = []
for fn in paired_df['filename']:
    img_path = next((p for p in sample_pngs if os.path.basename(p) == fn), None)
    if img_path:
        img = Image.open(img_path).resize((128, 128)).convert('RGB')
        X_images.append(np.array(img) / 255.0)
    else:
        X_images.append(np.zeros((128, 128, 3)))  # Fallback si pas match

X_images = np.array(X_images)
print("X_images shape:", X_images.shape)

# Labels + Split (stratify pour balance)
le = LabelEncoder()
y = le.fit_transform(paired_df['label'])
X_img_train, X_img_test, X_text_train, X_text_test, y_train, y_test = train_test_split(
    X_images, X_text, y, test_size=0.2, stratify=y, random_state=42
)
print(f"Split OK: Train classes: {len(np.unique(y_train))} | Test: {len(y_test)} samples")
print("Phase 3 Complète: Data prête pour modélisation (hybride texte + image). Valide: Shapes sans erreur ? Oui → Phase 4.")
```

```python
# Phase 4 Fixée : Imports + Resize + Baseline Simple
import tensorflow as tf
import matplotlib.pyplot as plt  # Fix NameError plt
import numpy as np

# Re-encode images à 224x224 (pour MobileNetV2, évite warning)
X_img_train_resized = np.array([tf.image.resize(img, [224, 224]).numpy() for img in X_img_train])
X_img_test_resized = np.array([tf.image.resize(img, [224, 224]).numpy() for img in X_img_test])
print("Images resized to 224x224: Train", X_img_train_resized.shape)

# Baseline Simple : Texte seul (Dense, pour debug – si acc low, problème data)
text_input = tf.keras.Input(shape=(X_text.shape[1],))
text_feat = tf.keras.layers.Dense(128, activation='relu')(text_input)
text_feat = tf.keras.layers.Dropout(0.3)(text_feat)
output_text = tf.keras.layers.Dense(len(le.classes_), activation='softmax')(text_feat)

model_text = tf.keras.Model(inputs=text_input, outputs=output_text)
model_text.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history_text = model_text.fit(X_text_train, y_train, epochs=10, batch_size=16, validation_split=0.2, verbose=1)
print(f"Baseline Texte Seul: Val Acc {history_text.history['val_accuracy'][-1]:.2%}")

# Hybride Full (texte + image, avec resize)
img_input = tf.keras.Input(shape=(224, 224, 3))
cnn = tf.keras.applications.MobileNetV2(input_tensor=img_input, weights='imagenet', include_top=False, pooling='avg')
cnn.trainable = False
img_feat = cnn(img_input)

text_input = tf.keras.Input(shape=(X_text.shape[1],))
text_feat = tf.keras.layers.Dense(128, activation='relu')(text_input)
text_feat = tf.keras.layers.Dropout(0.3)(text_feat)

combined = tf.keras.layers.Concatenate()([img_feat, text_feat])
output = tf.keras.layers.Dense(len(le.classes_), activation='softmax')(combined)

model = tf.keras.Model(inputs=[img_input, text_input], outputs=output)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit([X_img_train_resized, X_text_train], y_train, epochs=10, batch_size=16, validation_split=0.2, verbose=1)
print(f"Hybride Full: Val Acc {history.history['val_accuracy'][-1]:.2%}")

# Plot courbe (fix plt)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc Hybride')
plt.plot(history.history['val_accuracy'], label='Val Acc Hybride')
plt.title("Courbe Apprentissage Hybride")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history_text.history['accuracy'], label='Train Acc Texte Seul')
plt.plot(history_text.history['val_accuracy'], label='Val Acc Texte Seul')
plt.title("Courbe Baseline Texte")
plt.legend()
plt.show()

print("Debug Acc Low ? Check: 1. Labels uniques train {np.unique(y_train)}, 2. Texte sample {paired_df['text'].iloc[0]}, 3. Si <50%, itère Phase 3 (clean texte). Valide: Acc >50% baseline ? Oui → Phase 5.")
```

```python
# Debug Low Acc (Phase 4)
import numpy as np
from collections import Counter

print("=== Debug Data pour Low Acc ===")
print("1. Labels train uniques:", np.unique(y_train, return_counts=True))
print("Balance train:", Counter(y_train.tolist()))

print("2. Texte sample (top 3, check bruit 'patient report'):", paired_df['text'].iloc[0:3].tolist())

print("3. Labels sample:", paired_df['label'].iloc[0:3].tolist())

print("4. Nombre classes total:", len(le.classes_))
print("Si 1-2 classes ou texte boilerplate dominant, clean Phase 3.")
```

```python
# Itération Phase 3 : Clean Texte (Suppr Boilerplate)
import re
paired_df['text_clean'] = paired_df['text'].apply(lambda t: re.sub(r'patient report patient name classified date symptoms?', '', t).strip().lower() if 'patient report' in t else t)
# Suppr mots inutiles (new york hospital, etc.)
paired_df['text_clean'] = paired_df['text_clean'].apply(lambda t: re.sub(r'new york hospital|cant|bacause|hard respiratory movements?', '', t).strip())

print("Texte sample avant clean:", paired_df['text'].iloc[0:2].tolist())
print("Texte sample après clean:", paired_df['text_clean'].iloc[0:2].tolist())  # Ex. "my neck is annoying me sleep it"

# Re-encode TF-IDF avec clean
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_clean = TfidfVectorizer(max_features=300, stop_words='english')
X_text_clean = vectorizer_clean.fit_transform(paired_df['text_clean']).toarray()
print("X_text clean shape:", X_text_clean.shape)

# Re-split avec clean
from sklearn.model_selection import train_test_split
X_text_train_clean, X_text_test_clean, y_train, y_test = train_test_split(X_text_clean, y, test_size=0.2, stratify=y, random_state=42)
print("Clean OK – Use X_text_train_clean pour re-training.")
```

```python
# Re-Training Phase 4 avec Clean + Class Weight
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(enumerate(class_weights))
print("Class weights (pour balance):", {k: f"{v:.2f}" for k, v in class_weight_dict.items() if k < 5})  # Top 5

# Baseline Texte Seul (avec clean)
text_input = tf.keras.Input(shape=(X_text.shape[1],))
text_feat = tf.keras.layers.Dense(128, activation='relu')(text_input)
text_feat = tf.keras.layers.Dropout(0.3)(text_feat)
output_text = tf.keras.layers.Dense(len(le.classes_), activation='softmax')(text_feat)

model_text = tf.keras.Model(inputs=text_input, outputs=output_text)
model_text.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history_text = model_text.fit(X_text_train_clean, y_train, epochs=15, batch_size=16, validation_split=0.2, class_weight=class_weight_dict, verbose=1)
print(f"Baseline Texte Clean: Val Acc {history_text.history['val_accuracy'][-1]:.2%}")

# Hybride (avec clean texte + unfreeze CNN)
img_input = tf.keras.Input(shape=(224, 224, 3))
cnn = tf.keras.applications.MobileNetV2(input_tensor=img_input, weights='imagenet', include_top=False, pooling='avg')
cnn.trainable = True  # Unfreeze pour mieux apprendre
img_feat = cnn(img_input)

text_input = tf.keras.Input(shape=(X_text.shape[1],))
text_feat = tf.keras.layers.Dense(128, activation='relu')(text_input)
text_feat = tf.keras.layers.Dropout(0.3)(text_feat)

combined = tf.keras.layers.Concatenate()([img_feat, text_feat])
output = tf.keras.layers.Dense(len(le.classes_), activation='softmax')(tf.keras.layers.Dropout(0.5)(combined))

model = tf.keras.Model(inputs=[img_input, text_input], outputs=output)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit([X_img_train_resized, X_text_train_clean], y_train, epochs=15, batch_size=16, validation_split=0.2, class_weight=class_weight_dict, verbose=1)
print(f"Hybride Clean: Val Acc {history.history['val_accuracy'][-1]:.2%}")

# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Hybride Clean')
plt.plot(history.history['val_accuracy'], label='Val Hybride Clean')
plt.title("Courbe Hybride Clean")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history_text.history['accuracy'], label='Train Texte Clean')
plt.plot(history_text.history['val_accuracy'], label='Val Texte Clean')
plt.title("Courbe Baseline Clean")
plt.legend()
plt.show()

print("Valide: Val Acc >40% (vs 3% avant) ? Oui → Phase 5. Si low, ajoute samples Phase 3.")
```

```python
# Phase 5 Fixée : Évaluation Complète (Full Classes)
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Prédictions baseline texte clean
y_pred_t = np.argmax(model_text.predict(X_text_test_clean), axis=1)
acc_t = accuracy_score(y_test, y_pred_t)
f1_t = f1_score(y_test, y_pred_t, average='weighted')
print(f"Baseline Texte Clean: Acc {acc_t:.2%} | F1 {f1_t:.3f}")

# Prédictions hybride clean
y_pred_h = np.argmax(model.predict([X_img_test_resized, X_text_test_clean]), axis=1)
acc_h = accuracy_score(y_test, y_pred_h)
f1_h = f1_score(y_test, y_pred_h, average='weighted')
print(f"Hybride Clean: Acc {acc_h:.2%} | F1 {f1_h:.3f}")

# Rapport détaillé (full classes, pas slice)
print("\nRapport Baseline Texte Clean (Full 25 Classes):")
print(classification_report(y_test, y_pred_t, target_names=le.classes_, labels=range(len(le.classes_)), zero_division=0))

print("\nRapport Hybride Clean (Full 25 Classes):")
print(classification_report(y_test, y_pred_h, target_names=le.classes_, labels=range(len(le.classes_)), zero_division=0))

# Matrix baseline (top 10 classes pour visu)
cm_t = confusion_matrix(y_test, y_pred_t)
plt.figure(figsize=(10,8))
sns.heatmap(cm_t, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_[:10], yticklabels=le.classes_[:10])
plt.title("Confusion Matrix Baseline Texte Clean")
plt.xlabel("Prédit")
plt.ylabel("Vrai")
plt.show()

# Matrix hybride
cm_h = confusion_matrix(y_test, y_pred_h)
plt.figure(figsize=(10,8))
sns.heatmap(cm_h, annot=True, fmt='d', cmap='Reds', xticklabels=le.classes_[:10], yticklabels=le.classes_[:10])
plt.title("Confusion Matrix Hybride Clean")
plt.xlabel("Prédit")
plt.ylabel("Vrai")
plt.show()

# Test manuel (ex. symptôme "neck pain blurry vision headache")
sample_text = vectorizer_clean.transform(["neck pain blurry vision headache"]).toarray()
sample_img = np.expand_dims(X_img_test_resized[0], 0)
pred_sample_h = model.predict([sample_img, sample_text])
pred_disease_h = le.classes_[np.argmax(pred_sample_h)]
confidence_h = np.max(pred_sample_h)
print(f"Test Hybride 'neck pain blurry vision headache': Prédit '{pred_disease_h}' (confiance: {confidence_h:.2%})")

pred_sample_t = model_text.predict(sample_text)
pred_disease_t = le.classes_[np.argmax(pred_sample_t)]
confidence_t = np.max(pred_sample_t)
print(f"Test Baseline Texte: Prédit '{pred_disease_t}' (confiance: {confidence_t:.2%})")

print("Phase 5 Résumé: Baseline 30% Acc = Proto viable (texte seul pour chatbot). Hybride low = Overfit CNN. Next: Phase 6 (TFLite baseline) ou itère (+500 samples Phase 3).")
```

```python
# Phase 6 : Déploiement TFLite (Baseline Texte Seul)
converter = tf.lite.TFLiteConverter.from_keras_model(model_text)
converter.optimizations = [tf.lite.Optimize.DEFAULT]  # Quantise pour mobile
tflite_model = converter.convert()

with open('/content/chatbot_text_baseline.tflite', 'wb') as f:
    f.write(tflite_model)

# Taille + Test Inférence
import os
size_mb = os.path.getsize('/content/chatbot_text_baseline.tflite') / (1024**2)
print(f"TFLite Texte Seul: {size_mb:.1f} MB – Télécharge pour assets/ Android Studio.")

# Test TFLite
interpreter = tf.lite.Interpreter(model_content=tflite_model)
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Sample input (ex. texte clean "neck pain")
sample_text_tflite = X_text_test_clean[0:1].astype(np.float32)
interpreter.set_tensor(input_details[0]['index'], sample_text_tflite)
interpreter.invoke()
tflite_pred = interpreter.get_tensor(output_details[0]['index'])
tflite_disease = le.classes_[np.argmax(tflite_pred)]
tflite_conf = np.max(tflite_pred)
print(f"Test TFLite 'neck pain sample': Prédit '{tflite_disease}' (conf: {tflite_conf:.2%}) – Match Keras ?")

print("Phase 6 OK: TFLite prêt pour Kotlin (SymptomAnalyzer.kt). Next: Intégration mobile ou itère (+samples).")
```
